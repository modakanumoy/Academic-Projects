```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apples", "Banana","Cherries","Dates"]
myexplodes = [0.1,0.1,0.1,0.1]



plt.pie(y, labels = mylabels, startangle= 70, explode= myexplodes)
plt.legend(title = "list of fruites")
plt.show()
```
