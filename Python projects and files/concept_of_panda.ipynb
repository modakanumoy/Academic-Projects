```python
my_dataset = {'cars':['volvo','tyota','bmw','sujuki'],
              'passings':[3,4,8,9]
              }
my_var=pd.DataFrame(my_dataset)

print(my_dataset)
print(type(my_var))
print(my_var)
```

how to built a series 

```python
a = [3     ,        6,       7,         8,     10]
b = {'apple','oranges','grapes','pineapple','guava'}

my_var = pd.Series(a,b)

print(my_var)
print(type(my_var))
```

```python
calories = {'day 1': 500, 
            'day 2': 630,
            'day 3': 900
            }
my_var = pd.Series(calories)
print(calories)
print(my_var)
print(type(my_var))
```

```python
my_dataset1 = {'Days':['days1','days2','days3','days4'],
               'foods':['eggs','brocoli','black grams','chicken'],
               'quantity':[4,250,300,600],
                'units':['no.of eggs','grams','grams','grams']
            }
my_dataset2 = {'exercise_type': ['pushups','squts','chin-up','chest-press'],
               'no.of sets.': [4,5,6,9],
               'no.of reps': [10,12,12,15]
               }

my_var = pd.DataFrame(my_dataset1)
my_var2 = pd.DataFrame(my_dataset2)

print(my_dataset1)
print(my_dataset2)
print(my_var)
print(my_var2)
print(type(my_var))
print(my_var.loc[[1]])
print(my_var2.loc[[2]])
print(my_var.to_excel)
print(my_var2.to_excel)
print(my_var.to_string)
print(my_var2.to_string)

my_var.to_excel
my_var2.to_excel

dtx = pd.ExcelWriter("Data_analytics.xlsx")
my_var.to_excel(dtx)
dtx.save()

dtx2 = pd.ExcelWriter("Data_Analytics(2).xlsx")
my_var2.to_excel(dtx2)
dtx2.save()


read_data = pd.read_excel("Data_analytics.xlsx")
```
