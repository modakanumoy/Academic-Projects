Effect on Indian Share Market and Gold Due to Indian Budget 2024-25
Economic Analysis and Quantitative Data

Indian Share Market:

[1] Capital Expenditure Increase: The 17.6% increase in capital expenditure is likely to boost sectors like infrastructure, construction, and capital goods. Companies in these sectors are expected to see a positive impact on their stock prices.
[2] Reduction in Fiscal Deficit: The planned reduction in the fiscal deficit to 4.9% of GDP may lead to increased investor confidence, as it demonstrates fiscal prudence.
[3] Sectoral Allocations: Increased allocations to defense, health, education, and infrastructure sectors are expected to benefit companies operating in these areas.
[4] Tax Policies: Changes in tax policies can affect corporate profitability and investor sentiment. For instance, any reduction in corporate tax rates can lead to higher profits and potentially higher stock prices.

Gold Market:

[1] Inflation and Interest Rates: A significant increase in capital expenditure and revenue expenditure can lead to higher inflation. Gold is often seen as a hedge against inflation, so higher inflation expectations can drive up gold prices.
[2] Global Market Influence: Gold prices in India are also influenced by global gold prices and the value of the Indian rupee. A lower fiscal deficit can strengthen the rupee, which might reduce the local price of gold.
[3] Import Duties: Any changes in import duties on gold announced in the budget can directly affect gold prices in India. Lower import duties can reduce gold prices, while higher duties can increase them.

Changes in Policy Rates (2023–24)

| Policy Rate         | 2022-23 | 2023-24 | Change |
|---------------------|---------|---------|--------|
| Repo Rate           | 4.00%   | 4.50%   | +0.50% |
| Reverse Repo Rate   | 3.35%   | 3.75%   | +0.40% |
| CRR                 | 4.00%   | 4.25%   | +0.25% |
| SLR                 | 18.00%  | 18.25%  | +0.25% |
| MSF Rate            | 4.25%   | 4.75%   | +0.50% |


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
policy_rates = ['Repo Rate', 'Reverse Repo Rate', 'CRR', 'SLR', 'MSF Rate']
rate_2022_23 = [4.00, 3.35, 4.00, 18.00, 4.25]
rate_2023_24 = [4.50, 3.75, 4.25, 18.25, 4.75]

x = np.arange(len(policy_rates))  # label locations
width = 0.35  # width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, rate_2022_23, width, label='2022-23')
rects2 = ax.bar(x + width/2, rate_2023_24, width, label='2023-24')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Policy Rates')
ax.set_ylabel('Rates (%)')
ax.set_title('Policy Rates Comparison: 2022-23 vs 2023-24')
ax.set_xticks(x)
ax.set_xticklabels(policy_rates)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.show()
```
